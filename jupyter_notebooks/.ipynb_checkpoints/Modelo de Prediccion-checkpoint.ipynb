```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('marketing_campaign.csv', sep=';')
```

```python
# Creando el dataframe
df = pd.DataFrame(data,columns=data.columns)
df.head()
```

Variables:
+ **AcceptedCmp1** - 1 si el cliente aceptó la oferta en la primera campaña, 0 en caso contrario
+ **AcceptedCmp2** - 1 si el cliente aceptó la oferta en la segunda campaña, 0 en caso contrario
+ **AcceptedCmp3** - 1 si el cliente aceptó la oferta en la tercera campaña, 0 en caso contrario
+ **AcceptedCmp4** - 1 si el cliente aceptó la oferta en la cuarta campaña, 0 en caso contrario
+ **AcceptedCmp5** - 1 si el cliente aceptó la oferta en la quinta campaña, 0 en caso contrario
+ **Response (objetivo):** 1 si el cliente aceptó la oferta en la última campaña, 0 en caso contrario
+ **Complain**  - 1 si el cliente se quejó en los últimos 2 años
+ **DtCustomer:** fecha de inscripción del cliente en la empresa
+ **Education:** nivel de educación del cliente
+ **Marital:** estado civil del cliente
+ **Kidhome:** número de niños pequeños en el hogar del cliente
+ **Teenhome:** número de adolescentes en el hogar del cliente
+ **Income:** ingresos familiares anuales del cliente
+ **MntFishProducts:** cantidad gastada en productos pesqueros en los últimos 2 años
+ **MntMeatProducts:** cantidad gastada en productos cárnicos en los últimos 2 años
+ **MntFruits:** cantidad gastada en productos de frutas en los últimos 2 años
+ **MntSweetProducts:** cantidad gastada en productos dulces en los últimos 2 años
+ **MntWines:** cantidad gastada en productos vitivinícolas en los últimos 2 años
+ **MntGoldProds:** cantidad gastada en productos de oro en los últimos 2 años
+ **NumDealsPurchases:** número de compras realizadas con descuento
+ **NumCatalogPurchases:** número de compras realizadas utilizando el catálogo
+ **NumStorePurchases:** número de compras realizadas directamente en las tiendas
+ **NumWebPurchases:** número de compras realizadas a través del sitio web de la empresa
+ **NumWebVisitsMonth:** número de visitas al sitio web de la empresa en el último mes
+ **Recency:** número de días desde la última compra

## Regresión Logística

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
df_not_dummy = df;
df = df_w_dummy;
df["Response_Str"] = df["Response"].astype(str)
inv_unique = np.flip(df["Response_Str"].unique());
print(inv_unique)
plt.bar(inv_unique, df["Response_Str"].value_counts());
```

### Submuestreo

```python
#UnderSampling
df_no = df[df['Response'] == 0]
df_yes = df[df['Response'] == 1]
print("no_shape", df_no.shape)
print("yes_shape", df_yes.shape)
```

```python
df_no_reduced = df_no.sample(333,random_state=103)
df_no_reduced.shape
```

```python
df_reduced = pd.concat([df_no_reduced,df_yes],axis=0)
print(df_reduced.shape)
df_reduced.head()
```

```python
df_reduced.tail()
```

```python
df_reduced = df_reduced.sample(frac=1,random_state=103)
df_reduced.head()
```

```python
X = df_reduced['AcceptedCmp5'].values.reshape(-1, 1) # ['AcceptedCmp3','AcceptedCmp1','AcceptedCmp5']
y = df_reduced['Response']
# Creamos un split de datos
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=103)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver='lbfgs').fit(X_train, y_train)
y_train_hat = clf.predict(X_train)
y_test_hat = clf.predict(X_test)
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
pro = clf.predict_proba(X_train)[:,1]
pro
```

```python
plt.scatter(X_train,pro)
```

```python
pro = clf.predict_proba(X_test)[:,1]
plt.scatter(X_test,pro)
```

### Matriz de confusión

```python
from sklearn.metrics import confusion_matrix
labels = [1,0]
cm = confusion_matrix(y_test,y_test_hat,labels)
sns.heatmap(cm,
            annot=True,
            fmt='d',
            xticklabels=labels,
            yticklabels=labels,
            cmap='Greens')
plt.ylabel('Predicción')
plt.xlabel('Actual');
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test,y_test_hat))
print(classification_report(y_test,y_test_hat))
```

### Curvas ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score

plt.figure(figsize = (10, 6))
plt.plot([0,1], [0,1], 'r--')

probs = clf.predict_proba(X_test)

probs = probs[:, 1]
fpr, tpr, thresholds = roc_curve(y_test,probs)
roc_auc = roc_auc_score(y_test, y_test_hat)

label = 'AUC:' + ' {0:.2f}'.format(roc_auc)
plt.plot(fpr, tpr, c = 'g', label = label, linewidth = 4)
plt.xlabel('FPR', fontsize = 16)
plt.ylabel('TPR', fontsize = 16)
plt.title('ROC', fontsize = 16)
plt.legend();
```

Regla de dedo para un área bajo la curva en una curva ROC es:

- [0.5]: Equivalente a aleatorio.
- [0.5, 0.6): Test malo.
- [0.6, 0.75): Test regular.
- [0.75, 0.9): Test bueno.
- [0.9, 0.95): Test muy bueno.
- [0.95, 1): Muy bueno para ser cierto.

## Maquina de Vectores de Soporte

#### Normalizamos las variables

```python
from sklearn.preprocessing import MinMaxScaler
# Preparamos datos
df_var_ind = df_reduced.drop(["Response", "Response_Str"],axis=1)
X = df_var_ind.loc[:, best_variables]
y = df_reduced["Response"].values.reshape(-1, 1)

X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()
X_norm = X_scaler.fit_transform(X)
y_norm = y_scaler.fit_transform(y)
```

```python
X_norm.shape
```

```python
# Creamos un split de datos
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y_norm, test_size=0.33, random_state=103)
```

```python
X_train.shape
```

```python
# Creamos los modelos
from sklearn.svm import SVC

svc_rbf = SVC(kernel='rbf', C=100, gamma=0.1, probability=True)
svc_lin = SVC(kernel='linear', C=100, gamma='auto', probability=True)
svc_poly = SVC(kernel='poly', C=100, gamma='auto', degree=3, coef0=1, probability=True)
```

#### Aplicamos los modelos al conjunto de datos y graficamos la curva ROC

```python
import numpy as np

plt.style.use("seaborn")
plt.figure(figsize = (10, 6))
plt.plot([0,1], [0,1], 'r--')

print("Valores Score:")
print("============================================")
for name, svc, s in zip(["RBF","Linear","Poly"],[svc_rbf,svc_lin,svc_poly],["r","g","b"]):
    clf = svc.fit(X_train,y_train.ravel())
    y_train_hat = clf.predict(X_train)
    y_test_hat = clf.predict(X_test)
    if (name == "RBF"):
        clf_rbf = clf
        y_test_rbf = y_test
        y_test_hat_rbf = y_test_hat
    elif (name == "Linear"):
        clf_linear = clf
        y_test_linear = y_test
        y_test_hat_linear = y_test_hat
    elif (name == "Poly"):
        clf_poly = clf
        y_test_poly = y_test
        y_test_hat_poly = y_test_hat
    # Calculamos el error
    print(name, "Entrenamiento", clf.score(X_train, y_train))
    print(name, "Prueba", clf.score(X_test, y_test))
    print("============================================")
    probs = clf.predict_proba(X_test)
    probs = probs[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test,probs)
    roc_auc = roc_auc_score(y_test, y_test_hat)
    label = 'AUC ' + name + ': {0:.4f}'.format(roc_auc)
    plt.plot(fpr, tpr, c = s, label = label, linewidth = 4)
    
plt.xlabel('FPR', fontsize = 16)
plt.ylabel('TPR', fontsize = 16)
plt.title('ROC', fontsize = 16)
plt.legend();
```

#### Matrices de confusiones

```python
# RBF
from sklearn.metrics import confusion_matrix
labels = [1,0]
cm = confusion_matrix(y_test_rbf,y_test_hat_rbf,labels)
sns.heatmap(cm,
            annot=True,
            fmt='d',
            xticklabels=labels,
            yticklabels=labels,
            cmap='Greens')
plt.ylabel('Predicción')
plt.xlabel('Real');
print(accuracy_score(y_test_rbf,y_test_hat_rbf))
print(classification_report(y_test_rbf,y_test_hat_rbf))
```

```python
# Linear
from sklearn.metrics import confusion_matrix
labels = [1,0]
cm = confusion_matrix(y_test_linear,y_test_hat_linear,labels)
sns.heatmap(cm,
            annot=True,
            fmt='d',
            xticklabels=labels,
            yticklabels=labels,
            cmap='Greens')
plt.ylabel('Predicción')
plt.xlabel('Real');
print(accuracy_score(y_test_linear,y_test_hat_linear))
print(classification_report(y_test_linear,y_test_hat_linear))
```

```python
# Linear
from sklearn.metrics import confusion_matrix
labels = [1,0]
cm = confusion_matrix(y_test_poly,y_test_hat_poly,labels)
sns.heatmap(cm,
            annot=True,
            fmt='d',
            xticklabels=labels,
            yticklabels=labels,
            cmap='Greens')
plt.ylabel('Predicción')
plt.xlabel('Real');
print(accuracy_score(y_test_poly,y_test_hat_poly))
print(classification_report(y_test_poly,y_test_hat_poly))
```
