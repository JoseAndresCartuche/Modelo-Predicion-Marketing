## Regresión Logística

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
df["Response_Str"] = df["Response"].astype(str)
inv_unique = np.flip(df["Response_Str"].unique());
print(inv_unique)
plt.bar(inv_unique, df["Response_Str"].value_counts());
```

### Submuestreo

```python
df_no = df[df['Response'] == 0]
df_yes = df[df['Response'] == 1]
print("no_shape", df_no.shape)
print("yes_shape", df_yes.shape)
```

```python
df_no_reduced = df_no.sample(333,random_state=103)
df_no_reduced.shape
```

```python
df_reduced = pd.concat([df_no_reduced,df_yes],axis=0)
print(df_reduced.shape)
df_reduced.head()
```

```python
df_reduced.tail()
```

```python
df_reduced = df_reduced.sample(frac=1,random_state=103)
df_reduced.head()
```

```python
X = df_reduced['MntWines'].values.reshape(-1, 1)
y = df_reduced['Response']
# Creamos un split de datos
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=103)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver='lbfgs').fit(X_train, y_train)
y_train_hat = clf.predict(X_train)
y_test_hat = clf.predict(X_test)
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
pro = clf.predict_proba(X_train)[:,0]
```

```python
plt.scatter(X_train,pro)
```

### Matriz de confusión

```python
from sklearn.metrics import confusion_matrix
labels = [1,0]
cm = confusion_matrix(y_test,y_test_hat,labels)
sns.heatmap(cm,
            annot=True,
            fmt='d',
            xticklabels=labels,
            yticklabels=labels,
            cmap='Greens')
plt.ylabel('Predicción')
plt.xlabel('Actual');
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test,y_test_hat))
print(classification_report(y_test,y_test_hat))
```

### Curvas ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score

plt.figure(figsize = (10, 6))
plt.plot([0,1], [0,1], 'r--')

probs = clf.predict_proba(X_test)

probs = probs[:, 1]
fpr, tpr, thresholds = roc_curve(y_test,probs)
roc_auc = roc_auc_score(y_test, y_test_hat)

label = 'AUC:' + ' {0:.2f}'.format(roc_auc)
plt.plot(fpr, tpr, c = 'g', label = label, linewidth = 4)
plt.xlabel('FPR', fontsize = 16)
plt.ylabel('TPR', fontsize = 16)
plt.title('ROC', fontsize = 16)
plt.legend();
```
